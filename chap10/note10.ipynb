### 1. ipaddress 모듈
> 호스트 주소 객체 직접 생성

```python
import ipaddress

a4 = ipaddress.ip_address('192.0.2.1')
# IPv4Address('192.0.2.1')
print(a4)

a6 =  ipaddress.ip_address('2001:DB8::1')
# IPv6Address('2001:db8::1')
print(a6)

print(int(a4))
print(int(a6))


print(str(a4))
print(str(a6))

a4.is_global # False 출력됨, a4는 사설주소이다.
```

### 2. Socket 모듈
Socket이란? 
> 통신 채널의 <span style="color:red">종단점(end point)</span>이다. (<b>IP</b>와 <b>포트번호</b>로 식별)<br>
TCP인지 UDP인지 지정되어있다.<br>
호스트 컴퓨터의 이름이나 주소를 검색할 수 있는 함수가 내장되어있다.

```python
import socket
print(socket.gethostname())

host = socket.gethostname() 
print(socket.gethostbyname(host))
```

```python
import socket

# 문자열 이름
HOSTS = [ 
    'www.dongyang.ac.kr', 'pymotw.com',
    'www.python.org',
    'testsite',
]

for host in HOSTS:
    try:
        print('{} : {}'.format(host, socket.gethostbyname(host)))
    except socket.error as msg:
        print('{} : {}'.format(host, msg))
```
