```python
import telnetlib

HOST = "time.nist.gov"  # ì˜ˆì‹œ íƒ€ì„ ì„œë²„ (Daytime ì„œë¹„ìŠ¤ëŠ” ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©ë˜ì§€ ì•ŠìŒ)
PORT = 13  # Daytime í”„ë¡œí† ì½œ í¬íŠ¸

try:
    # 1. ì„œë²„ì— ì—°ê²°
    tn = telnetlib.Telnet(HOST, PORT, timeout=5)
    print(f"{HOST}:{PORT} ì— ì—°ê²°ë˜ì—ˆìŠµë‹ˆë‹¤.")
    
    # 2. ë°ì´í„° ìˆ˜ì‹  (Daytime ì„œë²„ëŠ” ì—°ê²° í›„ ë°”ë¡œ ì‹œê° ì •ë³´ë¥¼ ë³´ëƒ„)
    # read_all()ì€ EOF(ì—°ê²° ì¢…ë£Œ)ê¹Œì§€ ëª¨ë“  ë°ì´í„°ë¥¼ ì½ìŠµë‹ˆë‹¤.
    time_data_bytes = tn.read_all()
    time_data_str = time_data_bytes.decode('ascii').strip()

    from datetime import datetime
    # (1ë²ˆì˜ telnetlib ì—°ê²° ì½”ë“œì™€ time_data_str ìˆ˜ì‹  ì½”ë“œ ì´í›„ì— ì¶”ê°€)

    # ì˜ˆì‹œ: ìˆ˜ì‹ ëœ ì‹œê° ë¬¸ìì—´ (Daytime ì„œë²„ì˜ ì‹¤ì œ ì‘ë‹µ í˜•ì‹ì— ë§ì¶°ì•¼ í•¨)
    # ë§Œì•½ ì„œë²„ê°€ ì´ì™€ ê°™ì€ í˜•ì‹ìœ¼ë¡œ ë³´ë‚¸ë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
    # ì‹¤ì œ Daytime ì„œë²„ì—ì„œëŠ” ë³´í†µ 'YY-MM-DD HH:MM:SS' í˜•íƒœë¥¼ í¬í•¨í•©ë‹ˆë‹¤.
    # time_data_str = "55268 24-02-09 09:00:01 50 0 0 464.2 UTC(NIST) *"

    # ìˆ˜ì‹ ëœ ë¬¸ìì—´ì—ì„œ 'YY-MM-DD HH:MM:SS' ë¶€ë¶„ì„ ì¶”ì¶œí•˜ê±°ë‚˜, 
    # ì„œë²„ê°€ ì§ì ‘ ì‹œê° ë¬¸ìì—´ì„ ë³´ë‚¸ë‹¤ê³  ê°€ì •í•˜ê³  ì§„í–‰í•©ë‹ˆë‹¤.
    # ì—¬ê¸°ì„œëŠ” time_data_strì´ 'YY-MM-DD HH:MM:SS'ë¥¼ í¬í•¨í•œë‹¤ê³  ê°€ì •í•˜ê³ ,
    # ì‹¤ì œ ë°ì´í„° ì¶”ì¶œ ë¡œì§ì€ ìƒëµí•˜ê³  ì˜ˆì‹œ ê°’ìœ¼ë¡œ ëŒ€ì²´í•©ë‹ˆë‹¤.

    # ì‹¤ì œ ì„œë²„ ì‘ë‹µì„ íŒŒì‹±í•˜ì—¬ datetime ê°ì²´ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì´ í•„ìš”í•©ë‹ˆë‹¤.
    # ì˜ˆì‹œë¡œ, '24-02-09 09:00:01' ë¶€ë¶„ë§Œ ì¶”ì¶œë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
    date_time_part = "24-02-09 09:00:01" 

    # ìˆ˜ì‹ ëœ ì‹œê° ë¬¸ìì—´ì„ datetime ê°ì²´ë¡œ ë³€í™˜
    try:
        # '%y'ëŠ” ë‘ ìë¦¬ ì—°ë„, '%m'ì€ ì›”, '%d'ëŠ” ì¼, '%H'ëŠ” ì‹œ, '%M'ì€ ë¶„, '%S'ëŠ” ì´ˆ
        dt_object = datetime.strptime(date_time_part, '%y-%m-%d %H:%M:%S')
        
        # "2019 Feb 2 (Sat) 14:24:35" í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        # %Y: ë„¤ ìë¦¬ ì—°ë„, %b: ì›” ì•½ì–´, %d: ì¼, %a: ìš”ì¼ ì•½ì–´, %H: 24ì‹œ, %M: ë¶„, %S: ì´ˆ
        formatted_time = dt_object.strftime("%Y %b %d (%a) %H:%M:%S")
        
        print(f"âœ¨ ë³€í™˜ëœ í˜„ì¬ ì‹œê°: **{formatted_time}**")

    except ValueError as e:
        print(f"âŒ ì‹œê° ë³€í™˜ ì˜¤ë¥˜: ìˆ˜ì‹ ëœ ì‹œê° ë°ì´í„°ì˜ í˜•ì‹ì´ ì˜ˆìƒê³¼ ë‹¤ë¦…ë‹ˆë‹¤. {e}")
        
        # 3. ì—°ê²° ì¢…ë£Œ
        tn.close()
        
        print(f"ì„œë²„ì—ì„œ ìˆ˜ì‹ í•œ ì‹œê°: {time_data_str}")
    
except ConnectionRefusedError:
    print(f"ì—°ê²° ê±°ë¶€: ì„œë²„ {HOST}:{PORT} ì—ì„œ ì ‘ì†ì„ ê±°ë¶€í–ˆìŠµë‹ˆë‹¤. ì„œë¹„ìŠ¤ê°€ ì‹¤í–‰ ì¤‘ì¸ì§€ í™•ì¸í•˜ì„¸ìš”.")
except TimeoutError:
    print(f"íƒ€ì„ì•„ì›ƒ: ì„œë²„ {HOST}:{PORT} ì— ì—°ê²°í•˜ëŠ” ë° ì‹œê°„ì´ ì´ˆê³¼ë˜ì—ˆìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")
```

3. UDP í˜•íƒœë¡œ ì–´ë–»ê²Œ í•´ì•¼ í• ì§€ ìƒê°í•´ë³´ì•„ë¼

| êµ¬ë¶„ | TCP (Telnet/Daytime í”„ë¡œí† ì½œ) | UDP (NTP í”„ë¡œí† ì½œ) |
| :--- | :--- | :--- |
| **ì£¼ìš” ëª¨ë“ˆ** | `telnetlib` ë˜ëŠ” `socket` | **`socket`** |
| **í†µì‹  ë°©ì‹** | **ì—°ê²° ì§€í–¥í˜•** (Connection-Oriented) | **ë¹„ì—°ê²° ì§€í–¥í˜•** (Connection-less) |
| **ì‹ ë¢°ì„±** | ë°ì´í„° ì „ì†¡ì˜ **ì‹ ë¢°ì„± ë³´ì¥** (ì˜¤ë¥˜ ê²€ì‚¬, ì¬ì „ì†¡) | ì‹ ë¢°ì„± ë¹„ë³´ì¥ (íŒ¨í‚· ì†ì‹¤ ê°€ëŠ¥) |
| **ì£¼ìš” í¬íŠ¸** | Daytime: **13ë²ˆ** | NTP: **123ë²ˆ** |
| **ë°ì´í„° ì²˜ë¦¬** | ì„œë²„ì— ì—°ê²° í›„, ì„œë²„ê°€ ë°ì´í„°ë¥¼ ì „ì†¡í•˜ë©´ ìˆ˜ì‹  | í”„ë¡œí† ì½œ ê·œê²©ì— ë§ëŠ” **ìš”ì²­ íŒ¨í‚·**ì„ ì‘ì„±í•˜ì—¬ ì „ì†¡ í›„, ì‘ë‹µ íŒ¨í‚· ìˆ˜ì‹  |
| **ì†ë„** | ì—°ê²° ì„¤ì •(3-way handshake) ê³¼ì •ìœ¼ë¡œ ì¸í•´ ìƒëŒ€ì ìœ¼ë¡œ ëŠë¦¼ | ì—°ê²° ì„¤ì • ê³¼ì •ì´ ì—†ì–´ **ìƒëŒ€ì ìœ¼ë¡œ ë¹ ë¦„** |

```python
import socket
import struct
from datetime import datetime, timedelta

NTP_SERVER = 'pool.ntp.org' # ì „ ì„¸ê³„ ê³µìš© NTP ì„œë²„ í’€
NTP_PORT = 123               # NTP í¬íŠ¸ (UDP)
NTP_PACKET_SIZE = 48         # NTP íŒ¨í‚· í¬ê¸°ëŠ” ê³ ì • 48ë°”ì´íŠ¸
# NTP Epoch (ì‹œê°„ ê¸°ì¤€)ëŠ” 1900ë…„ 1ì›” 1ì¼ UTC
NTP_EPOCH = datetime(1900, 1, 1) 
# Unix Epoch (1970ë…„ 1ì›” 1ì¼)ì™€ NTP Epoch ê°„ì˜ ì‹œê°„ ì°¨ì´ (ì´ˆ ë‹¨ìœ„)
UNIX_EPOCH = datetime(1970, 1, 1)
NTP_TO_UNIX_DELTA = (UNIX_EPOCH - NTP_EPOCH).total_seconds() 

def get_ntp_time():
    """NTP ì„œë²„ì—ì„œ ì‹œê°„ì„ ê°€ì ¸ì™€ ì§€ì •ëœ í˜•ì‹ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤."""
    
    # 1. UDP ì†Œì¼“ ìƒì„± ë° íƒ€ì„ì•„ì›ƒ ì„¤ì •
    try:
        client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        client_socket.settimeout(5)
        print(f"âœ… {NTP_SERVER}:{NTP_PORT} ì— ì—°ê²° ì‹œë„ ì¤‘...")
    except Exception as e:
        print(f"âŒ ì†Œì¼“ ìƒì„± ì˜¤ë¥˜: {e}")
        return

    # 2. NTP ìš”ì²­ íŒ¨í‚· ìƒì„± ë° ì „ì†¡
    # NTP ìš”ì²­ íŒ¨í‚·: 48ë°”ì´íŠ¸. ì²« ë²ˆì§¸ ë°”ì´íŠ¸(LI/VN/Mode)ë§Œ ì„¤ì •í•˜ê³  ë‚˜ë¨¸ì§€ëŠ” 0.
    # '\x1b' (ì‹­ì§„ìˆ˜ 27): 
    #   LI=0 (ê²½ê³  ì—†ìŒ), VN=3 (NTPv3), Mode=3 (í´ë¼ì´ì–¸íŠ¸ ëª¨ë“œ)
    ntp_packet = b'\x1b' + (NTP_PACKET_SIZE - 1) * b'\x00' 
    
    try:
        client_socket.sendto(ntp_packet, (NTP_SERVER, NTP_PORT))
        
        # 3. ì‘ë‹µ ìˆ˜ì‹ 
        data, addr = client_socket.recvfrom(1024)
        client_socket.close()
        
    except socket.timeout:
        print(f"âŒ íƒ€ì„ì•„ì›ƒ: ì„œë²„ {NTP_SERVER}:{NTP_PORT} ì—ì„œ ì‘ë‹µì„ ë°›ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")
        client_socket.close()
        return
    except Exception as e:
        print(f"âŒ ë°ì´í„° ì†¡ìˆ˜ì‹  ì˜¤ë¥˜: {e}")
        client_socket.close()
        return

    # 4. ì‹œê° ì •ë³´ ì¶”ì¶œ ë° ë³€í™˜
    try:
        # NTP ì‘ë‹µ íŒ¨í‚·ì—ì„œ 'Transmit Timestamp' (43ë²ˆì§¸ ë°”ì´íŠ¸ë¶€í„° 8ë°”ì´íŠ¸)ë¥¼ ì¶”ì¶œ
        # ì´ íƒ€ì„ìŠ¤íƒ¬í”„ëŠ” 64ë¹„íŠ¸ ì •ìˆ˜(ì •ìˆ˜ë¶€ 32ë¹„íŠ¸ + ì†Œìˆ˜ë¶€ 32ë¹„íŠ¸)ë¡œ êµ¬ì„±ë¨
        # struct.unpack('>I', ...)ëŠ” ë¹… ì—”ë””ì–¸(Big-Endian, >) ë°©ì‹ì˜ ë¶€í˜¸ ì—†ëŠ” 4ë°”ì´íŠ¸ ì •ìˆ˜(I)ë¡œ ë³€í™˜í•˜ë¼ëŠ” ì˜ë¯¸
        ntp_time_integer = struct.unpack('>I', data[40:44])[0]
        
        # NTP Time (1900ë…„ ê¸°ì¤€ ì´ˆ)ì„ UNIX Time (1970ë…„ ê¸°ì¤€ ì´ˆ)ìœ¼ë¡œ ë³€í™˜
        unix_time = ntp_time_integer - NTP_TO_UNIX_DELTA
        
        # Unix Timeì„ datetime ê°ì²´ë¡œ ë³€í™˜
        current_time_utc = datetime.utcfromtimestamp(unix_time)
        
        # ë¡œì»¬ ì‹œê°„ìœ¼ë¡œ ë³€í™˜ (ì‹œìŠ¤í…œì˜ í˜„ì¬ ì‹œê°„ëŒ€ ì„¤ì • ì‚¬ìš©)
        current_time_local = datetime.fromtimestamp(unix_time)
        
        # "2019 Feb 2 (Sat) 14:24:35" í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        formatted_time = current_time_local.strftime("%Y %b %d (%a) %H:%M:%S")
        
        print(f"ğŸ’¡ NTP ì„œë²„ ({addr[0]})ì—ì„œ ìˆ˜ì‹ í•œ ì‹œê°:")
        print(f"   - UTC ì‹œê°: {current_time_utc.strftime('%Y-%m-%d %H:%M:%S')} (NTP) ")
        print(f"   - ë³€í™˜ëœ ì‹œê°: **{formatted_time}**")
        
    except Exception as e:
        print(f"âŒ ì‹œê° ë³€í™˜ ì˜¤ë¥˜: íŒ¨í‚· ì²˜ë¦¬ ì¤‘ ë¬¸ì œ ë°œìƒ. {e}")
        
if __name__ == "__main__":
    get_ntp_time()
```
